Research Question:
"How does food insecurity in low-income communities correlate with obesity rates and health outcomes across different regions in the United States?"

Peer-Reviewed Sources:

Walker, R. E., Keane, C. R., & Burke, J. G. (2010). Disparities and Access to Healthy Food in the United States: A Review of Food Deserts Literature. Health & Place.
Hilmers, A., Hilmers, D. C., & Dave, J. (2012). Neighborhood Disparities in Access to Healthy Foods and Their Effects on Environmental Justice. American Journal of Public Health.

Problem Statement
Food insecurity disproportionately impacts low-income and minority communities, often leading to higher obesity rates and poorer health outcomes due to limited access to nutritious foods. This study examines USDA’s Food Access Research Atlas and CDC’s Obesity & Chronic Disease Data to assess whether limited access to grocery stores in low-income communities is a strong predictor of obesity prevalence.

Hypothesis: Low-income areas with limited food access will show higher obesity rates compared to areas with better food availability.

USDA Food Access Research Atlas
Source: U.S. Department of Agriculture

```python
# Fill missing numerical values with the median
food_access_data.fillna(food_access_data.median(), inplace=True)
health_outcome_data.fillna(health_outcome_data.median(), inplace=True)

# Fill missing categorical values with the mode
food_access_data.fillna(food_access_data.mode().iloc[0], inplace=True)
health_outcome_data.fillna(health_outcome_data.mode().iloc[0], inplace=True)
```

```python
food_access_data.rename(columns={'LILATracts_1And10': 'Low_Access_1_Mile', 'LILATracts_halfAnd10': 'Low_Access_Half_Mile'}, inplace=True)
health_outcome_data.rename(columns={'Obesity_AdjPrev': 'Obesity_Prevalence'}, inplace=True)
print(food_access_data.columns)
print(health_outcome_data.columns)
```

```python
# Creating a new column for food desert severity
food_access_data['Food_Desert_Severity'] = food_access_data.apply(
    lambda row: 3 if row['Low_Access_1_Mile'] == 1 and row['Low_Access_Half_Mile'] == 1 
    else (2 if row['Low_Access_1_Mile'] == 1 else (1 if row['Low_Access_Half_Mile'] == 1 else 0)), axis=1)

# Display first few rows to verify new column
food_access_data[['Low_Access_1_Mile', 'Low_Access_Half_Mile', 'Food_Desert_Severity']].head()
```
