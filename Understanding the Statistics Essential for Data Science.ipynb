## Descriptive Statistics
Descriptive Statistics is the method of organizing, summarizing, and presenting data in an informative way.  This could include measures of statistics, charts, and graphs.  

### Measures of Central Tendency
Is an average of all the values in the data set a single numerical value around all the values seem to cluster.  Facilitate the comparison of data and allows informed decision making.  

The mathematical average (**mean** value) of a data set is calculated by finding the sum of all the values and dividing by the total number of values.  Mathematical averages work better when all the values are evenly distributed and is affected by extreme values.  The mean is not ideal for data sets with highly disparate data.  

The Positional Average is obtained by first ordering the values as ascending or descending order and identifying the value in the middle.  This is the **median** value for the data set. In certain situations the median value is better as it is less affected by extreme values. 

1. **Arithmetic Mean**.  The ratio of the sum of all values to the total number of observations.  

2. **Geometric Mean**.  a mean or average which indicates a central tendency of a finite set of real numbers by using the product of their values.

3. **Harmonic Mean**. The harmonic mean is one of the three Pythagorean means. For all positive data sets containing at least one pair of nonequal values, the harmonic mean is always the least of the three means, while the arithmetic mean is always the greatest of the three and the geometric mean is always in between. Since the harmonic mean of a list of numbers tends strongly toward the least elements of the list, it tends (compared to the arithmetic mean) to mitigate the impact of large outliers and aggravate the impact of small ones. 

4. **Median**.  The middle value or observation in a given set of data.  If the observations are even in number then the value at the middle of the range such that half of the values are avobe the median and half of the values are below the median. 

5. **Mode**.  Qualitative analysis that seeks to display the most frequently occuring value or observation.  

6. **Quartile**.  The range of the data ordered ascending or descending that is divided into four even sections.  25% of values are below Q1 and 75% of values are above quartile 1.  Q2 divides the range into fifty percent of the values above and fifty percent of the values are below Q2.  **The Q2 value is the median.**  Q3 represents 25% of values above Q3 and 75% of values are below Q3.

7.  **Decile**.  The range of the data ordered as ascending or descending and divided into ten equal parts.  Each group is a decile.  Each part represents 1/10th of the data.

8.  **Percentile**.  The range of the data ordered ascending or descending divided into 100 equal parts and is represented as a percent value ranging from 1 to 100.  This can be expressed as decimal values or numbers with a percent sign. The value indicates the percentage of data that falls below the stated value.




### Measures of Dispersion
Dispersions are the measure of variation in values in a data set.  An index that quantifies the degree to which the data is dispersed or spread.  Serves as a reliability of the average.  The smaller the dispersion, the more reliable the average as it demonstrates a closer grouping with fewer extreme values.

### Range Outliers, Quartile Deviation
Measures of dispersion include the range, outliers, quartile deviations.
1. Range. The range is the measure of the spread of the data from the maximum to the minimum value.  The range can be quantified by subtracting the maximum value and minimum value.  The range is affected by outliers or extreme values. The range therefore does not make it a good indicator of dispersion.

2. Outliers.  Extreme values or observations that can skew the data toward larger values or smaller values.  The presence of extreme values affects the range giving the appearance of a mean within the range but affected by the presence of extreme values.  Outliers also skew quartile and decile values and are not representative of the majority of values or observations.

3.  Quartile deviation.  Another measure of dispersion that seeks to identify values that are outliers from the quartile ranges.  The quartile range is calculated as one half the difference of Q3 and Q1.  Values outside the quartile deviation are outliers.

### Mean Absolute Deviation, Standard Deviation, Variance
1. Mean Absolute Deviation.  The average of absolute deviations is called the mean deviation of each value from the arithmetic mean. 

2. The Standard Deviation.  The measure of the spread of data from the average.  This is a calculated value.  The standard deviation magnifies larger deviations and a larger spread is captured and is the most used measure of dispersion.

3. Variance.  The measurement of how far a data set is dispersed from the mean or average value.  This is calculated as the squared vale of the standard deviation.  Since each term is squared, the variance is always positive or zero.  

### Z-score, Standard Score, Empirical Rule
1. Z-score.  The z-score is known as the standard score and is a calculation of how many deviations a particular data value is above or below the arithmetic mean.   Values above the arithmetic mean have positive Z-scores while those below the arithmetic mean are negative values.

2. Empirical Rule.  Uses ordered data ascending or descending to create a bell shaped curve or normal distribution of the values.  This designates how much of the data lies within one, two and three standard deviations.  At the center, 65% of the data have Z-scores between zero and one standard deviation from the mean.  95% of the data is within two standard deviations, and 99.7% of the data is within three standard deviations.  Values outside of three standard deviations should be examined for potential as extreme values.  Not all large distributions from the mean are outliers.

### Coefficient of Variation
This is a calculated value to determine the amount of variation in a data set.  The data set with a smaller coefficient of variation is more reliable data set and less dispersed across the range.

### Measures of Shapes
Skewness is a measure to describe the shape of a data set that can be represented graphically.  
Normal distribution is a graphical display of the data in which the mean, median, and mode are equal.  In this distribution 50% of the data lies above and the remaining 50% lies below the arithmetic mean.  The data is evenly distributed across the range and is symmetrical.

**Positively Skewed**. The data set has a frequency curve with a long tail toward the right, nearer the larger extreme values.  The mode is to the left, the median will be between the mode and the mean which trends right with the extreme values. 

**Negatively Skewed**.  The data set has a frequenct curve with a long tail toward the left, nearer smaller extreme values.  The mode is toward the right and the median is positioned between the mode and the mean.  The arithmetic mean trends to the left, toward the smaller extreme values.

**Coefficient of Skewness**  A calculated value that skews either left or right by the positive or negative coefficient value.

## Probability
Uncertainty exists in every decision made in society.  The ability to make informed decisions with confidence uses probability.  Probability is the study of uncertainty helping to assess potential outcomes for planning or decision making; the likelihood of an event occuring.

Probability states that the likelihood of an event occurring is equal to the ratio of the number of favorable outcomes to the total number of outcomes.  Probability is used in machine learning algortithms to determine if an email is spam or not.

The relationship between events are mutually exclusive when they cannot occur simultaneously.  Totally exclusive, here at leas one of the events will occur but does not impact the other possible outcomes.  The set of all possible outcomes is a **sample space**.

1. Classical approach.  When there is the probability of an event A where P(A) in n number of times is mutually exclusive, equally likely outcomes with k as outcomes is k/n.
2. Relative Approach.  The probability of an event is the fraction of times it has occurred in several experiments. Past data may be used in lieu of experimental trials.
3. Subjective Approach.  Besides past events, the probability of events are estimated using the subjective judgement of experts.  

### Conditional Probability
Very often the likelihood of an event depends on the possibility of other events occurring.  Obtaining relevant information can give that will affect the likelihood of the initial event.  Conditional probability is the measure of the likelihood of an event occurring given the condition of another event that has already occurred.

### Independent and Dependent Events
**Complement** of an event A is the set of outcomes in the sample space that are not in the event.  Complements are mutually exclusive and exhaustive in nature, including all other outcomes that are not desired as an outcome for A.  

**Independent Event**: Events are independent when the occurrance (nor not) of one does not impact the occurrance of the other.  Implies there are an equal number of opportunities for each outcome and one is not favored over another.

**Dependent Event**: The occurrance of events (or not) of the first event affects the likelihood of the subsequent events.  The likelihood of any following events changes with each event such that the following events occurring are changed by the previous event.

### Addition Theorem of Probability
With all events in a series being mutually exclusive (independent) events, the probability that at least one event will occur is the sum of their probabilities.  

When those events are not mutually exclusive the union of A and B is equal to the probability of event A added to the probability of B minus the probability of A and B.

### Multiplication Theorem of Probability
the probability of two events, A and B, is equal to the probability of AB multiplied by the probability of B.  

For Independent events that are mutually exclusive, the likelihood of A and B occurring is equal to the product of their individual probabilities.  

### Bayes Theorem
The conditional probability of an event, based on the occurrence of another event, is equal to the likelihood of the second event  given the first event multiplied by the probability of the first event.

## Sampling and Sampling Techniques
Large units of population makes it uneconomical to approach the entire population.  It is therefore necessary to evaluate a representative sample of the population.  The sample statistics can then be used to evaluate the population.  To carefully evaluate the population it is necessary to realize there are a variety of samples that can represent different aspects of the population each producing varied results. Sample collection is important and the possibility of errors must be considered when selecting the sample.

**Sampling** is the process of making inductive inferences about an entire population based on a sample.

Sampling errors are errors in the inductive inference from a sample to a population.  any other error is a non-sampling error.  Non-sampling errors can include data collection errors, biases in responses from customers.  Nonsampling errors occur in both sampling studies and complete enumeration.  

Sampling is adventageous as it is the only option when destructive testing is involved.  Census is costly and time-consuming, sampling provides reasonable assurance.  Study or examination of all units is often cumbersome and error prone due to monotony, fatigue, and other factors.  The sampling process has a significant impact on a large mass of data.  

Sampling is limited by changes of bias, difficulty in selecting a sample in a population, exclusion in sampling impacts accuracy when compared back to the population.

### Probability Sampling Methods
Probability or random sampling is a sampling approach where the investigator selects the sample based on the principles of probability.

1. Simple Random Sampling: A technique of sample selection in which every sample of a given size has an equal probability of inclusion in the sample and equa likely outcome.

2. Systematic Sampling: Ann alternative approach when the selection of a sample are cumbersome.  Systematic sampling is a technique in which samples are selected from a sequentially arranged lot, at regular intervals.  The value k is called the sampling interval.  Systematic sampling offers scope for convienence in sample selection.  Sometimes units need not be referenced numerically.

3. Stratified random sampling: The sampling procedure using a heterogenous population that are segmented into homogenous layers called strata.  A simple random sample is then taken from each homogeneous strata.  The samples selected are a better representation of the population rather than random sampling in a heterogeneous population.  

4. Cluster sampling: The population is divided into clusters that are equally heterogeneous.  A cluster is selected using stratified or simple random sample techniques.  A sample of clusters is representative of the population as they are equally heterogeneous and serves as a powerful tool when geographic dispersion is wide.  The reuslts are poor when the sampled clusters do not represent the population.

### Non-probability Sampling Methods
An approach where the sample is selected based on the subjective decision of the investigator.  There are four non-probability sampling techniques:
1. Convienence Sampling: Sample units are selected entirely based on convienence or convienent approach rather than randomly. (customer satisfaction survey following visit within days of visit)
2. Voluntary Response Sampling: Individuals comprise the population and some volunteer to participate in the study as sample respondents. This process of obtaining small populations of those who choose to respond likely have intention or incentive to answer.
3. Judgemental or purposeful sampling: The investigator selects units from the population using subject matter expertise regarding the population.  This is a targeted population to determine specialized information from particular populations.
4. Snowball sampling: Individuals are the population units and following response recommend their participation to others as referrals to complete respondence for the study.  This can be incentive based but can offer niche information for specialized populations that are difficult to cultivate.

### Null and Alternate Hypotheses
Two types of hypotheses are suggested when designing a scientific study.  Beginning with two hypotheses:
1. **Null hypothesis**: The negation of the statement under test.
2. **Alternative hypothesis**: The statement under test.

### P value
The p-value is used to test the outcome of hypotheses.  The p-value assumes the data assumes a normal distribution and there is no change in the standard deviation of the data.  The level of significance is denoted as alpha and is selected to give a specific level of confidence to the resulting answer.  The p-value is an indicator of the ease with which you can reject the null hypothesis.  When p-value is low the null can be easily rejected.  When the p-value is higher, the null is more difficult to reject as it has become a significant probability.

### Levels of Significance
This is an alternative approach to determine the level of significance, alpha.  This is simply stated that if a level of significance is set at a particular value, any values higher than that are accepted and values lower than the alpha will be rejected.  At lower values the rejection becomes more difficult .  There is a level of significance at which it is too low or too high to reject the null hypothesis, this value is the p-value.  


### Type I and Type II Errors
Hypothesis testing outcomes can generate errors when the hypothesis is incorrectly rejected or accepted.  The decsion and outcomes of hypothesis testing are prone to two error types:
1. **Type I Error**: The Null Hypothesis is true and it is rejected.  Denoted as alpha
2. **Type II Error**: Null Hypothesis is false and it is accepted. Denoted as beta.  

Usually the value of alpha is set as 0.05 or 0.001 as the level of significance representing 5% and 1% level of significance.  alpha and beta are never more than one but the sum of alpha and beta does not equal one.  If alpha or beta is equal to 1 the other is automatically zero.

### Z-Test

The threshold limit is set using python or other programming languages.  Calculation of the value is using standard normal tables which is stored in programming languages.

Simultaneous study of two variables and the analysis of the data observed or produced.  This allows the establishment of a statistical relationship between the variables and subsequent study of the relationship between the variables.

### Parametric vs. Non-parametric Tests
Statistical testing of hypothesis can be divided as:
1. Parametric Tests:  As the name suggests, this is tests based on particular parameters that have a measurable or calculable outcome.  Data conforming to probability distributions can be used as a historical context for present day data as a statistical distribution of sufficient size.  Tests against particular features or parameters in a sample size are parametric tests.  
2. Non-parametric Tests: These are statistical inferences made from categorical data that have a qualitative response.

### Test of Significance
Multiple populations can also be compared using tests of significance.  This is calculated with sufficient sample size to approach the normal distribution.  The sample sizes do not need to be the same but each sample size should be of a significantly high number.

### Z-Test
Parametric test used the population proportion and population sample mean to determine the distribution of data. When null is true Z follows the standard normal distribution and can be calculated using the data table or software such as R and Python.  The p-value is compared to the z-score and the determination to accept or reject the null hypothesis. 

### T-Test
When the population standard deviation is unknown, the t-test is used.  The t-test is used when independent samples  are chosen from two data sets.  The difference between the means of the two independent samples are chosen is the basis for inference between the two samples.  **Paired T-test** is used when samples are dependent, inferences can be drawn between one individual value with multiple or historical data points.  In such cases the data is assembled into pairs and the assumption is the sample size will be equal.  Missing or null values would be excluded if one or more data points were missing from the set.   

### Parametric Tests, ANOVA
Analysis of Variance (ANOVA) is used to compare the means of several populations that have been segmented for homogeneity.  The analysis of variance serves as a powerful technique to allow statistical comparision between the populations.  ANOVA enables analysis if variations between populations are due to intrinsic factors.  The relative difference in the population mean and the variance is a measure of when to reject or support the null hypothesis.  When the population mean is close to the variance the null hypothesis is held to be true.  When there is considerable distance between the population mean and the variance it indicates the null hypothesis is false.

### Chi-Square Test
Non-parametric tests do not require any assumptions of a statistical distribution.  Chi square is a non-parametric test for independence or dependence between variables.  Tabulated data can be examined for probability distributions as a relative frequency.  Attributes are independent when the observed frequency has a marginal difference between the expected frequency.  Dependence is noted when the value for Chi Square is large.  

**Fisher's Exact Test** is a discrete distribution that is a hypergeometric distribution that is an improvement measure of success over the chi-square test.  When 20% or more of the values have expected frequencies that are less than five the chi-square value fails.   

### Sign Test
The paired t-test is used to test whether the mean of two distributions are equal when paired samples are used.  The sign test is the non-parametric test version of paired t-test.  The Sign test looks at the change in proportion between two attributes as a basis of inference between two samples.  When a large enough sample is used the t-test can be used and is more accurate because it demonstrates magnitude as well as the direction indicated in the Sign Test.

### Kruskal Wallis Test
An alternative to the one way ANOVA whrn the requisites to perform an ANOVA are not satisfied.  This tests the significance in difference observed for sample means.  The sample size and values to infer from the data.  This test does not require normal distribution or sufficient sample size is not available.  

### Mann Whitney Wilcoxon Test
Independent random samples from two populations can be compared even when the sample sizes are not equal using the Mann-Whitney Wilcoxon Test.  The population is segmented to compare characteristics based on feature attributes that are ranked and sorted by rank based on frequency.  When there are significant difference based on the threshold previously set, the hypothesis is rejected.

### Run Test for Randomness
Run tests are logical tests that can guard against incorrect conclusions or biases.  These tests examine the data for true randomness and is expecially important when using random number generators.  A sequence of identical occurrences that are proceeded or followed by another occurrence.  When data occur in the non-random manner the data is sufficiently large or too small.  These threshold limits are based on the levels of significance.

## Relation Between Variables
The relationship between variables are examined as correlation between the variables as compared to causation.  These can be calculated using specific coeeficients that give information about the directionality and magnitude of relationships.  

The scatter diagram indicates a relationship through the formation of a band using the plotting of the raw data.  the directionality of the slop indicates direction with upwards sloping that both variables are increasing and downward sloping indicates a decrease in one variable as the other increases.  The width of the bands also demonstrates the strength of the relationship and can be calculated to determine how far each point is from a mean value.  Tighter, more narrow spread in the scatter indicates a stronger relationship between variables as compared to a wider range.

Scatter plots also permit the visualization of linear or curvilinear relationships as relative shape and direction of the banding pattern.  Linear relationships indicate incremental change of a variable while curvilinear relationships indicate a more advanced relationship such as exponential or logrithmic relationships.

The rank correlation coefficient assigns ranks to scores as an ascending or descending sorting order. ### Causation
The relationship between two variables where one is the cause and the other is a measure of the effect on the first variable.  Usually the cause is denoted as X and the effect is variable Y.  The correlation coefficient quantifies the magnitude or strength of the relationship.  Correlation may not be indicative of a cause and effect between the two relationship, it can be the observation or spurious correlation.  Correlation is carefully studied as causation using regression equations.  

### Regression
A plot of the data in a scatter plot can show a visual relationship between the variables.  Calculating the mean, standard deviation, and correlation can be calculated for a two variable data set in which one variable is an independent variable or cause and the second is a dependent variable or measured effect.  A prediction value can then be determined from the regression line and the values can be extrapolated for future predictions and values not observed can be calculated using the resulting equation from the regression.  Regressions can take any shape although this chapter only uses linear regression. 

### Coefficient of Determination
The coefficient of determination is an index to estimate the quality of the fit of the regression equation.  When the regression is used to validate the effectiveness of a prediction it is imperative the regression error is calulated to improve confidence of the inferences made from the regression.

### Quantifying Quality
When the quality of the prediction is good the calculations for the error terms are small.  The coeffieicnt of determination specifically quantifies the variations in data points aligning to the regression equation.  When the variation in data is small, the fit to the regression equation is good, the value for the coefficient of determination is higher, as close to 1 as possible.  Two highly correlated variables with a regression equation providing quality prediction, the unexplained variation will be low.

### How to not lie with Statistics
The techniques are well understood and the assumptions about the tools being used are well understood.  Incorrect approaches include: 
1. Drawing strong conclusions rather than making inferences from hypothesis testing specifically when small sample sizes are used.  The threshold limit will be higher when the sample size is small.  In general the findings from analysis need to be carefully communicated as hypothesis testing does not validate an assertion but implies that there is insufficient evidence.
2. Using purposive sampling over random sampling where the results are exclusively sampled from those with an interest in the results.  Convienenece sampling can exclude useful data from those who have yet to take an interest.  
The possibility of errors in data is higher when the sampling is delegated and a lack of understanding for the questions.

#### Lying Through Visualizations
Charts serve as powerful aid to communicate information on data sets.  When presented visually data must consist of the correct data, scale, and correct type of chart. Incorrect projections can be the result of planned distortion or biases of the investigator. 

#### Lying About Relationships
Not all correlation is from cause and effect.  correlations are sometimes false, therefore predictions of the outcome of another variable should be avoided.  Correct regression equations should be selected based on the data type and linear or curvilinear relatedness of the data points.  When there is a causation relationship the cause should be properly determined to be used for prediction.  The sample used cannot have relationship determined merely by relatedness of the data.  The correlation coefficient and level of significance should be determined to confidently states the relationship between variables.   Coeffieicnt of determination is an index that quantifies the quality of an equation for predictive use.  Calculation of the significance must be performed.  If the test of significance does not reject a null hypothesis the equation cannot be used for predictive capacity.
